```python
# prompt: 2. Check the given number is prime or not.

def is_prime(n):
  """
  Checks if a given number is prime.

  Args:
    n: The number to check.

  Returns:
    True if the number is prime, False otherwise.
  """
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

# Get input from the user
num = int(input("Enter a number: "))

# Check if the number is prime and print the result
if is_prime(num):
  print(num, "is a prime number")
else:
  print(num, "is not a prime number")
```

```python
def find_factors(number):
    factors = []
    for i in range(1, number + 1):
        if number % i == 0:
            factors.append(i)
    return factors

num = int(input("Enter a number: "))

factors = find_factors(num)
print(factors)

if len(factors) == 2:
    print("The number is prime")
else:
    print("The number is not prime")
    for i in factors:
        print("The factors of", num, "are:", i)
```

```python
import numpy as np # Imports the NumPy library for numerical operations, aliased as 'np'.
import matplotlib.pyplot as plt # Imports the 'pyplot' module from Matplotlib for plotting, aliased as 'plt'.

plt.axes(projection='polar') # Creates a polar axes object for plotting in polar coordinates.
r = 3 # Sets the radius of the circle to 3.
rads = np.arange(0, (2 * np.pi), 0.01) # Creates an array of angles from 0 to 2π with a step of 0.01.

# plotting the circle
for i in rads: # Iterates through each angle in 'rads'.
  plt.polar(i, r, 'g.') # Plots a point in polar coordinates with angle 'i', radius 'r', and a green dot marker.

plt.show() # Displays the complete plot with the circle.
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)  # Create an array of theta values from 0 to 2π
r = 5 * (1 + np.cos(theta))  # Calculate the corresponding r values for the cardioid

plt.polar(theta, r, 'r')  # Plot the cardioid in polar coordinates with a red line
plt.title('Cardioid: r = 5(1 + cos(theta))')  # Set the title of the plot
plt.show()  # Display the plot

```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 1000)  # Create an array of theta values from 0 to 2π
r = 2 * np.abs(np.cos(2 * theta))  # Calculate the corresponding r values for the rose

plt.polar(theta, r, 'r')  # Plot the rose in polar coordinates with a red line
plt.title('Four-Leaved Rose: r = 2|cos(2θ)|')  # Set the title of the plot
plt.show()  # Display the plot
```

```python
import numpy as np  # Imports the NumPy library for numerical operations
import matplotlib.pyplot as plt  # Imports the Matplotlib library for plotting

plt.axes(projection='polar')  # Sets up a polar coordinate system for the plot
a = 3  # Defines a variable 'a' and assigns it the value 3
rad = np.arange(0, (2 * np.pi), 0.01)  # Creates an array of angles from 0 to 2π

# Loops through each angle in the 'rad' array
for i in rad:
    r = a + (a * np.cos(i))  # Calculates the radius for the first cardioid
    plt.polar(i, r, 'g.')  # Plots the first cardioid in green

    r1 = a - (a * np.cos(i))  # Calculates the radius for the second cardioid
    plt.polar(i, r1, 'r.')  # Plots the second cardioid in red

plt.show()  # Displays the plot
```
